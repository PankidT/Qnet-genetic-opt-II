# Pareto front

```python
def plot_pareto_frontier(Xs, Ys, maxX=True, maxY=True):
    '''Pareto frontier selection process'''
    sorted_list = sorted([[Xs[i], Ys[i]] for i in range(len(Xs))], reverse=maxY)
    pareto_front = [sorted_list[0]]
    for pair in sorted_list[1:]:
        if maxY:
            if pair[1] >= pareto_front[-1][1]:
                pareto_front.append(pair)
        else:
            if pair[1] <= pareto_front[-1][1]:
                pareto_front.append(pair)            
    
    '''Plotting process'''
    plt.scatter(Xs,Ys, label='Data')
    pf_X = [pair[0] for pair in pareto_front]
    pf_Y = [pair[1] for pair in pareto_front]
    plt.plot(pf_X, pf_Y, 'x-', color='red', label='Optimal Pareto')
    plt.xlabel("1 - Fidelity")
    plt.ylabel("Cost")
    plt.title("Pareto frontier")
    plt.legend(loc="upper right", fontsize=13)
    plt.show()

    return pf_X, pf_Y
```

```python
path = 'Results/Results_17_10_2023 (only default node)'
with open(f'{path}/result_validate_exper_2023-10-10 15:50:09.pickle', "rb") as f:
    data = pickle.load(f)

inverse_fidelity = []
for data_point in data.exper_config.FidelityHistory.All[-1]:
    inv_F = 1 - data_point
    inverse_fidelity.append(inv_F)

# plt.scatter(data.exper_config.FidelityHistory.Mean, data.exper_config.CostHistory.Mean)
# plt.scatter(inverse_fidelity, data.exper_config.CostHistory.Mean)

pX, pY = plot_pareto_frontier(inverse_fidelity, data.exper_config.CostHistory.All[-1], maxX=False, maxY=False)

print(f'Pareto X: {pX}')
print(f'Pareto Y: {pY}')
print(f'Fidelity {[1-i for i in pX]}')
```
